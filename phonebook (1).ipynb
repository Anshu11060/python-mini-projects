```python
# empty dictionary 
contact = {}

contact["ANSHU"] = {"phone":"9958608510", "email": "anshusingh@gmail.com"}
contact["ATUL"] = {"phone": "7292098352", "email": "atulsingh@gmail.com"}
contact["ADI"] = {"phone": "9910951030", "email" : "adsingh@gmail.com"}
```

```python
def add_contact():
    name = input("enter your name")
    if name.upper() in contact:
        print(f"conatact name {name} already exists")

    else:
        phone = input("enter you phone no.-->")
        email = input("enter your e-mail-->")
        contact[name] = {"phone": phone, "email": email}
        print(f' contact for {name}  added successfully\n')
```

```python
#VIEW CONTACT
def view_contact():
    if not contact:
        print("No contacts available.")
    else:
        for name, info in contact.items():
            print(f"Name: {name}")
            print(f"Phone: {info.get('phone', 'N/A')}")
            print(f"Email: {info.get('email', 'N/A')}\n")
```

```python
#search contact
def search_contact():
    search = input("Enter name to search: ").upper()
    if search in contact:
        print("-------- CONTACT FOUND --------")
        print(f"Name: {search}")
        print(f"Phone: {contact[search]['phone']}")
        print(f"Email: {contact[search]['email']}\n")
    else:
        print("Contact not found.")
```

```python
#update contact
def update_contact():
    name = input("enter name in which you want to change").upper()
    if name in contact:
        phone = input("enter  updated phone no.--->")
        email= input("enter  updated email---->")
        contact[name]= {'phone': phone,'email': email}
    else:
        print("contact not found")
```

```python
#delete contact
def delete_contact():
    name = input("Enter name to delete: ").upper()
    if name in contact:
        del contact[name]
        print(f"Contact {name} deleted successfully.\n")
    else:
        print("Contact not found.")
```

```python
while True:
    print("-----------------PHONEBOOK-------------------")
    print("1 ADD CONTACT \n 2. VIEW ALL CONTACTS \n 3. SEARCH CONTACT \n 4. UPDATE CONTACT \n 5. DELETE CONTACT \n 6.EXIT  ")
    choice = input("enter your choice ")

    if choice == "1":
        add_contact()
    elif choice == "2":
        view_contact()
    elif choice == "3":
        search_contact()
    elif choice == "4":
        update_contact()
    elif choice == "5":
        delete_contact()
    elif choice == "6":
         print("thanks for using . GOODBYE")
         break

    else:
        print("invalid choice")

    
        
```
