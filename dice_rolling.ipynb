dice rolling project

```python
import random
```

```python
user_input = input("PRESS Y TO ROLL THE DICE")
while user_input == "y":
  no = random.randint(1,6)


if no == 1:
    print("[-----]")
    print("[     ]")
    print("[  0  ]")
    print("[     ]")
    print("[-----]")


if no == 2:
  print("[-----]")
  print("[ 0   ]")
  print("[     ]")
  print("[     ]")
  print("[   0 ]")
  print("[-----]")

if no == 3:
  print("[-----]")
  print("[     ]")
  print("[ 000 ]")
  print("[     ]")
  print("[     ]")
  print("[-----]")

if no == 4:
  print("[-----]")
  print("[     ]")
  print("[ 0  0]")
  print("[     ]")
  print("[ 0  0]")
  print("[-----]")

if no == 5:
  print("[-----]")
  print("[     ]")
  print("[ 0  0]")
  print("[  0  ]")
  print("[ 0  0]")
  print("[-----]")

if no == 6:
  print("[-----]")
  print("[     ]")
  print("[ 000 ]")
  print("[ 000 ]")
  print("[ 000 ]")
  print("[-----]")

user_input = input("PRESS Y TO ROLL THE DICE and anything else to exit ---->")
print("/THANKS FOR ROLLING")
```
